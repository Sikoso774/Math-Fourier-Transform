Introduction à la transformée de Fourier discrète

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, ifft
```

Exemple de Stem Plot

```python
# Génération des données
N = 10  # Nombre de points
n = np.arange(N)  # L'axe des x (index des échantillons)
x = np.sin(2 * np.pi * n / N)  # Le signal discret (par exemple, une sinusoïde)

# Création du graphique avec plt.stem
plt.stem(n, x, basefmt=" ")

# Personnalisation des labels et du titre
plt.xlabel('Index (n)')
plt.ylabel('Amplitude')
plt.title('Exemple de graphique Stem Plot')

# Affichage du graphique
plt.grid(True)
plt.show()
```

Exercice 2

```python
# Définition du Dirac

def dirac_delta(n, position=0):
  """
  Génère une impulsion de Dirac discrète.

  Args:
    n: Un tableau NumPy représentant les indices.
    position: La position où l'impulsion est égale à 1 (par défaut 0).

  Returns:
    Un tableau NumPy représentant l'impulsion de Dirac.
  """
  return np.where(n == position, 1, 0)

# Génération des données
N = 16  # Nombre de points
n = np.arange(N)  # L'axe des x (index des échantillons)

# Générer l'impulsion de Dirac centrée à n=0
impulsion = dirac_delta(n, position=0)

# Création du graphique avec plt.stem
plt.stem(n, impulsion, basefmt=" ")

# Personnalisation des labels et du titre
plt.xlabel('Index (n)')
plt.ylabel('Amplitude')
plt.title('Impulsion de Dirac Discrète')

# Affichage du graphique
plt.grid(True)
plt.show()
```

```python
# Représentation de la TFD du Dirac en O

# Génération des données
N = 16  # Nombre de points
n = np.arange(N)  # L'axe des x (index des échantillons)
x = np.fft.fft(dirac_delta(n,position=0))  # La TFD du Dirac

# Création du graphique avec plt.stem
plt.stem(n, x, basefmt=" ")

# Personnalisation des labels et du titre
plt.xlabel('Index (n)')
plt.ylabel('Amplitude')
plt.title('La Transformée de Fourier discrète du Dirac')

# Affichage du graphique
plt.grid(True)
plt.show()
```

```python
# Génération des données
N = 16  # Nombre de points
n = np.arange(N)  # L'axe des x (index des échantillons)

# Générer l'impulsion de Dirac centrée à n=11
impulsion = dirac_delta(n, position=11)

# >>>>> Une impulsion du dirac en n = 11

# Création du graphique avec plt.stem
plt.stem(n, impulsion, basefmt=" ")

# Personnalisation des labels et du titre
plt.xlabel('Index (n)')
plt.ylabel('Amplitude')
plt.title('Impulsion de Dirac Discrète')

# Affichage du graphique
plt.grid(True)
plt.show()
```

```python
# Représentation de la TFD du Dirac en 11

# Génération des données
N = 16  # Nombre de points
n = np.arange(N)  # L'axe des x (index des échantillons)
x = np.fft.fft(dirac_delta(n, position=11))  # La TFD du Dirac
TFD_exct = np.exp((-2j * np.pi * n * 11) / N)

# Création du graphique avec plt.stem
plt.stem(n, x, basefmt=" ", label="TFD Calculée")  

# Création de la courbe de TFD_exct
plt.plot(n, TFD_exct, label="TFD Exacte")

# Personnalisation des labels et du titre
plt.xlabel('Index (n)')
plt.ylabel('Amplitude')
plt.title('La Transformée de Fourier discrète du Dirac')

# Ajout de la légende
plt.legend()

# Affichage du graphique
plt.grid(True)
plt.show()
```

```python
# Représentation du signal discret x1
k  = 3
N  = 12
n  = np.arange(N)
x1 = (np.exp(2j * np.pi*k*n/N)+np.exp(-2j *np.pi*k*n/N))/2

# Représentation du signal discret x2
x2 = (np.exp(2j * (2*np.pi*k*n))-np.exp(-2j * (2*np.pi*k*n)))/2j
```

```python
# TFD de x1 et x2
y1= np.fft.fft(x1)
y2= np.fft.fft(x2)

# Création du graphique
plt.stem(n, y1, basefmt=" ", label="TFD de x1")
plt.plot(n,y1) 
#plt.stem(n, y2, basefmt= "", label="TFD de x2")
plt.title("Les TFD obtenues")
plt.legend()
plt.grid(True)
plt.show()
```
